```python
def guiyihua(X):
#     return (X - X.min()) / (X.max() - X.min())
    return (X - X.mean()) / X.std()
```

```python
a = np.array([80.6000,
77.9000,
75.5000,
72.6000
])  #真实值
a
```

```python
b = np.array([67.8437,
84.1349,
76.9862,
72.9898
])  #拟合值
```

```python
plt.figure(figsize=(8,4))
plt.plot(np.arange(2012, 2016), a, label='监测数据真实值')
plt.plot(np.arange(2012, 2016), b, linestyle='--', label='综合舆情拟合值')
plt.xlim(2012, 2015)
plt.xlabel('年份', fontsize=14)
plt.ylabel('氨氮排放量（万吨）', fontsize=14)
plt.xticks(range(2012, 2016))
plt.ylim(0, 100)
# plt.title('新浪微博用户发展趋势')
plt.legend(loc=2)
# plt.grid(True)
ax2 = plt.twinx()
ax2.plot(np.arange(2012, 2016), abs(c*100), ':r', label='绝对百分比误差（%）')
ax2.set_ylabel('误差(%)')
ax2.set_xticks(range(2012, 2016))
ax2.legend()
ax2.set_ylim(0,100)
# plt.show()
```

```python
x = ['2011年', '2012年', '2013年', '2014年', '2015年']
y = [82.7, 80.6, 77.9, 75.5, 72.6]
plt.bar(x=x, height=y, label='', color='rbg', alpha=0.8)
# 在柱状图上显示具体数值, ha参数控制水平对齐方式, va控制垂直对齐方式
for x1, yy in zip(x, y):
    plt.text(x1, yy + 1, str(yy), ha='center', va='bottom', fontsize=10, rotation=0)
# 设置标题
# plt.title("历年水环境质量")
# 为两条坐标轴设置名称
plt.ylim(0,90)
plt.xlabel("\n年份", fontsize=14)
plt.ylabel("农业源氨氮排放量（万吨）\n", fontsize=14)
# 显示图例
# plt.legend()
```

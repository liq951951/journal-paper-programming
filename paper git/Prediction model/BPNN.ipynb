```python
import pandas as pd
import  numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import optimizers
from sklearn.model_selection import KFold
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from keras.optimizers import SGD, Adam, RMSprop
import openpyxl
```

```python
adam = optimizers.adam(lr=0.001)
# average = []
temp= []
kf = KFold(n_splits=10)
```

10折交叉验证

```python
def MODEL(read_path,feature_dim):
    model = Sequential()
    model.add(Dense(4,activation='relu',input_dim=feature_dim))
    # model.add(Dense(4, activation='relu'))
    model.add(Dense(1,activation='sigmoid'))
    model.compile(optimizer=adam,
                  loss='mse',
                  metrics=['accuracy'])
    data = pd.read_excel(read_path,sheet_name = 'Sheet1')
    print(data)
#     sum = []
    for train_index, test_index in kf.split(data.values):
        X_train = data.iloc[train_index, 2:].values
        y_train = data.iloc[train_index, 2].values
        X_test = data.iloc[test_index, 2:].values
        y_test = data.iloc[test_index, 2].values
        model.fit(X_train,y_train)
        pred = model.predict(X_test)
        y_pred = np.array(pred).reshape(-1).tolist()
        print("预测情况：",y_pred)
        print("真实情况：",y_test)
        MAE = mean_absolute_error(y_test, y_pred)
        MSE = mean_squared_error(y_test, y_pred)
        temp.append((MAE, MSE))
        
#         sum.append(pred)
#     sum = np.array(sum)
#     average.append(sum.mean(axis=0))
```

随机取样

```python
def MODEL(read_path,feature_dim):
    model = Sequential()
    model.add(Dense(4,activation='relu',input_dim=feature_dim))
    # model.add(Dense(4, activation='relu'))
    model.add(Dense(1,activation='sigmoid'))
    model.compile(optimizer=adam,
                  loss='mse',
                  metrics=['accuracy'])
    data = pd.read_excel(read_path,sheet_name = 'Sheet1')
    print(data)
    train_data = data.iloc[:,2:]
    train_target = data.iloc[:,2]
    X_train,X_test, y_train, y_test = train_test_split(train_data,train_target,test_size=0.2, random_state=111)
    model.fit(X_train,y_train)
    pred = model.predict(X_test)
    y_pred = np.array(pred).reshape(-1).tolist()
    y_test = np.array(y_test).reshape(-1).tolist()
    print("预测情况：",y_pred)
    print("真实情况：",y_test)
    MAE = mean_absolute_error(y_test, y_pred)
    MSE = mean_squared_error(y_test, y_pred)
    loss = [MAE, MSE]
    return loss
    
        
#         sum.append(pred)
#     sum = np.array(sum)
#     average.append(sum.mean(axis=0))
```

```python
loss = MODEL(r"E:\水环境\论文修改\数据\河水污染对压力关键词.xlsx",15)
print('MAE, MSE = ', loss)
```

```python
def MODEL1(read_path,feature_dim):
    model = Sequential()
    model.add(Dense(5,activation='relu',input_dim=feature_dim))
    # model.add(Dense(4, activation='relu'))
    model.add(Dense(1,activation='sigmoid'))
    model.compile(optimizer= RMSprop(0.001),
                  loss='mse',
                  metrics=['accuracy'])
    data = pd.read_excel(read_path,sheet_name = 'Sheet1')
    print(data)
    y_train = data.iloc[:-12,1]
    x_train = data.iloc[:-12,2:]
    y_test = data.iloc[-12:,1]
    x_test = data.iloc[-12:,2:]
    model.fit(x_train,y_train, epochs=1000)
    pred = model.predict(x_test)
    pred = np.array(pred).reshape(-1).tolist()
    y_pred = np.array(pred).reshape(-1).tolist()
    y_test = np.array(y_test).reshape(-1).tolist()
    print("预测情况：",y_pred)
    print("真实情况：",y_test)
    con = pd.DataFrame([y_pred, y_test])
    
    MAE = mean_absolute_error(y_test, y_pred)
    MSE = mean_squared_error(y_test, y_pred)
    loss = [MAE, MSE]
    return loss, con
```

```python
loss1, con1 = MODEL1(r"E:\水环境\论文修改\数据\河水污染对压力关键词.xlsx",15)
print('MAE, MSE = ', loss1)
```

```python
writer = pd.ExcelWriter(r'E:\水环境\论文修改\数据\BPNN预测结果.xlsx',engine='openpyxl')
book = openpyxl.load_workbook(writer.path)
writer.book = book
con1.to_excel(excel_writer=writer,sheet_name='河水污染',encoding='utf-8')
writer.close()
```

```python
def guiyihua(X):
#     return (X - X.min()) / (X.max() - X.min())
    return (X - X.mean()) / X.std()
```

```python
a = np.array([1153.8,
1125.8,
1102.4,
1068.6
])  #真实值
b = np.array([974.4812,
1217.2454,
1124.2972,
1073.8117
])  #拟合值
c = (b- a) / a
```
